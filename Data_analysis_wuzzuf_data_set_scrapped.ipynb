```python
from bs4 import BeautifulSoup

import csv 

import requests

from time import sleep

from random import randint

import pandas as pd

import numpy as np
```

```python

title = []
location = [] 
published = []
level = []
time = []
company = []
hyper_link = []
```

```python


headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36 Edg/108.0.1462.54",
           "Accept-Encoding": "gzip, deflate, br",
           "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
           "DNT":"1",
           "Connection":"close",
           "Upgrade-Insecure-Requests": "1"}

# doing np.arange to deal with the pagination and be able to scrape all data analysis jobs 
pages = np.arange(0, 11)

for page in pages:
    
    link = 'https://wuzzuf.net/search/jobs/?a=navbl&q=data%20analyst&start='+str(page)
    wuzzuf_link = requests.get(link, headers = headers)
    soup = BeautifulSoup(wuzzuf_link.content, 'html.parser')
    data_analyst = soup.find_all('div' , class_= "css-pkv5jc")
    sleep(randint(2,8))
   
        
    
    for job in data_analyst:
          
        description_div = soup.find('div', class_= 'css-y4udm8')
        
        job_title = job.find('h2', class_= 'css-m604qf').text
        title.append(job_title)
        
        job_location = job.find('span', class_= 'css-5wys0k').text
        location.append(job_location)
        
        job_published = job.find('div', class_= 'css-do6t5g')
        
        
        job_level = description_div.find('a', class_= 'css-o171kl').text
        level.append(job_level)
        
        job_time = description_div.a.span.text
        time.append(job_time)
        company_name = job.find('a', class_= 'css-17s97q8').text
        company.append(company_name)
        job_link = job.find('a', class_= 'css-o171kl' )['href']
        hyper_link.append(job_link)
        
        if job_published is not None:
            published.append(job_published.text.strip(',ago'))       
                 
            print(f'''              
Title: {job_title}
Company name: {company_name}
Career Level: {job_level}
Job type: {job_time}
Location: {job_location}
Published: {job_published.text.strip(',ago')}
Link: https://wuzzuf.net{job_link}''' )
        
```

```python
 
data = {'Title': title,
'Company_name': company,
'Career_Level': level,
'Job_type': time,
'Location': location,
'Link': f'''https://wuzzuf.net{job_link}''',
#'Published': published     
 }
df = pd.DataFrame(data)
df_published = pd.DataFrame({'Published': published})

# done a Concat because the data frame giving me (ValueError: All arrays must be of the same length)
wuzzuf_data_analysis_data_set = pd.concat([df, df_published],axis=1)
wuzzuf_data_analysis_data_set
```
